Three Dimensional Reflection

Reflect the point (0,0,1) through the XY-plane.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.transform import Rotation as R
import math
def Refz(v):
 r=R.from_matrix([[1,0,0],[0,1,0],[0,0,-1]])
 return r.apply(v)

ax=plt.axes(projection='3d')
ax.scatter(0,0,1) 

a=Refz([0,0,1])
print(a)

ax.scatter(a[0],a[1],a[2]) 
plt.show()


```
