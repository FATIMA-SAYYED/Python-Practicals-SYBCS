Q.1. Generate 3-D contour plots for the following functions.
	f(x,y)=sin⁡(x+y)for x,y ∈[-2π,2π]


```python
from mpl_toolkits import mplot3d
from pylab import *
import numpy as np
def f(x,y):
  return np.sin(x+y)
x=np.linspace(-2*pi,2*pi,100)
y=np.linspace(-2*pi,2*pi,100)
X,Y= np.meshgrid(x,y)
Z=f(X,Y)
ax=axes(projection='3d')
ax.contour3D(X,Y,Z,50)
xlabel('x')
ylabel("y")
title("3D Contour")
legend()
show()
```

Q.2. Generate 3-D surface plots for the following functions.
	f(x,y)=x^2+y^2,-6≤x,y≤6


```python
from mpl_toolkits import mplot3d
from pylab import *
import numpy as np
def f(x,y):
  return x**2+y**2
x=np.linspace(-6,6,30)
y=np.linspace(-6,6,30)
X,Y= np.meshgrid(x,y)
Z=f(X,Y)
ax=axes(projection='3d')
ax.plot_surface(X,Y,Z)
xlabel('x')
ylabel("y")
title("3D Surface plot")
legend()
show()
```

Q.3. Generate 3-D wireframe plots for the following functions.
	f(x,y)= e^(x^2+y^2 )  for x,y ∈[0,2π]


```python
from mpl_toolkits import mplot3d
from pylab import *
import numpy as np
def f(x,y):
  return np.exp(x**2+y**2)
x=np.linspace(0,2*pi,30)
y=np.linspace(0,2*pi,30)
X,Y= np.meshgrid(x,y)
Z=f(X,Y)
ax=axes(projection='3d')
ax.plot_wireframe(X,Y,Z,rstride=4,cstride=4)
xlabel('x')
ylabel("y")
title("3D wireframe plot")
legend()
show()
```

	f(x,y)=x.e^(〖-x〗^2-y^2 ) for x,y∈[-6,6]

```python
from mpl_toolkits import mplot3d
from pylab import *
import numpy as np
def f(x,y):
  return np.exp((-x)**2-y**2)
x=np.linspace(-6,6,50)
y=np.linspace(-6,6,50)
X,Y= np.meshgrid(x,y)
Z=f(X,Y)
ax=axes(projection='3d')
ax.plot_wireframe(X,Y,Z,rstride=3,cstride=3)
xlabel('x')
ylabel("y")
title("3D wireframe plot")
legend()
show()
```

```python
from mpl_toolkits import mplot3d
from pylab import *
import numpy as np
def f(x,y):
  return np.sin(x+y)+cos(x)
x=np.linspace(-2*pi,2*pi,50)
y=np.linspace(-2*pi,2*pi,50)
X,Y= np.meshgrid(x,y)
Z=f(X,Y)
ax=axes(projection='3d')
ax.plot_wireframe(X,Y,Z,rstride=2,cstride=2)
xlabel('x')
ylabel("y")
title("3D wireframe plot")
legend()
show()
```
