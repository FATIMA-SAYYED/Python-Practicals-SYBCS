Q.1. write a python program to draw a line y = x in [-5,5] with suitable label in the X-axis, Y-axis and a title.

```python
from pylab import *
import numpy as np
x=np.linspace(-5,5,100)
y=x
plot(x,y,label="$y=x$")
xlabel('x')
ylabel("y")
title("Graph of $y=x$")
legend()
show()
```

Q.2. Plot the graph of f(x) =sin(x) and g(x) =cos(x) in [-2π,2π] (Legend at lower left corner)

```python
from pylab import *
import numpy as np
x=np.linspace(-2*pi,2*pi,100)
f=np.sin(x)
g=np.cos(x)
plot(x,f,label="$sin(x)$")
plot(x,g,label="$cos(x)$")
xlabel('x')
ylabel("y")
title("Graph of $f(x) =sin(x)$ and $g(x) =cos(x)$")
legend(loc=3)
show()
```

Q.3. Plot the graph of f(x) = 1/log(x) in [-5,5] with line width=5 and in green color.

```python
from pylab import *
import numpy as np
x=np.linspace(-5,5,100)
y=(1/np.log(x))
plot(x,y,"g",lw=5,label="$y=1/logx$")
xlabel('x')
ylabel("y")
title("Graph of $y=1/logx$")
legend()
show()
```

Q.4. Plot the graph of f(x) = e^(-0.4x).sin⁡x in [0,4π] with red dashed line with circle markers.

```python
from pylab import *
import numpy as np
x=np.linspace(0,4*pi,20)
y=np.exp(-0.4*x)*np.sin(x)
plot(x,y,"--or",label="$y=e^(-0.4x).sinx$")
xlabel('x')
ylabel("y")
title("Graph of $y=e^(-0.4x).sinx$")
legend()
show()
```

Q.5. Generate a subplots for the following function using python.
	y=x,y=x^2,y=x^3,y=x^4,[-5,5]  with 2×2 subplots

```python
from pylab import *
import numpy as np
from math import *
x=np.linspace(-5,5,10)
y1=x
y2=x**2
y3=x**3
y4=x**4
subplot(2,2,1)
plot(x,y1,label="$y1=x$")
legend()
subplot(2,2,2)
plot(x,y2,label="$y2=x^2$")
legend()
subplot(2,2,3)
plot(x,y3,label="$y3=x^3$")
legend()
subplot(2,2,4)
plot(x,y4,label="$y4=x^4$")
legend()
show()
```

```python
from pylab import *
import numpy as np
from math import *
x=np.linspace(-5,5,10)
y1=1/x
y2=1/x**2
y3=1/x**3
subplot(1,3,1)
plot(x,y1,label="$y1=1/x$")
legend(loc=4)
subplot(1,3,2)
plot(x,y2,label="$y2=1/x^2$")
legend(loc=4)
subplot(1,3,3)
plot(x,y3,label="$y3=1/x^3$")
legend(loc=4)
show()
```

	y=x^2 sin⁡(1/x),y=x^2 sin⁡(1/x^2 ),[0,4π]  with 2×1 subplots

```python
from pylab import *
import numpy as np
from math import *
x=np.linspace(0,4*pi,10)
y1=x**2*np.sin(1/x)
y2=x**2*np.sin(1/x**2)
subplot(2,1,1)
plot(x,y1,label="$y1=x^2.sin⁡(1/x)$")
legend()
subplot(2,1,2)
plot(x,y2,label="$y2=x^2.sin⁡(1/x^2)$")
legend()
show()
```

Plot the graph of 2 polynomials in interval [-10, 10]
	f(x)=3.x^4+2.x^3+7.x^2+2.x+9 and g(x)=5.x^3+9.x+2
	With font size=20, color=blue & green, choose x, y values correspondingly.

```python
from pylab import *
import numpy as np
x=np.linspace(-10,10,100)
f=3*x**4+2*x**3+7*x**2+2*x+9
g=5*x**3+9*x+2
plot(x,f,label="$3.x^4+2.x^3+7.x^2+2.x+9$",color="blue")
plot(x,g,label="$5.x^3+9.x+2$",color="green")
text(7,20000,"$3.x^4+2.x^3+7.x^2+2.x+9$",fontsize=20,color="blue")
text(7.5,5000,"$5.x^3+9.x+2$",fontsize=20,color="green")
xlabel('x')
ylabel("y")
legend()
show()
```
