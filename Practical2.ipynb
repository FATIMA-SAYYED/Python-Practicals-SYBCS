3D Contour Points

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

def f1(x,y):
  return np.exp(-x**2-y**2)

x=np.linspace(-2,2)
y=np.linspace(-2,2)
X,Y=np.meshgrid(x,y)
Z=f1(X,Y)
ax=plt.axes(projection='3d')
ax.contour3D(X,Y,Z,50)
```

Wireframe Plot


```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

def f1(x,y):
  return np.exp(-x**2-y**2)

x=np.linspace(-2,2)
y=np.linspace(-2,2)
X,Y=np.meshgrid(x,y)
Z=f1(X,Y)
ax=plt.axes(projection='3d')
ax.plot_wireframe(X,Y,Z,color='green')
```

Surface Plot

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

def f1(x,y):
  return np.exp(-x**2-y**2)

x=np.linspace(-2,2)
y=np.linspace(-2,2)
X,Y=np.meshgrid(x,y)
Z=f1(X,Y)
ax=plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,cmap='Greens')
```

Surface Triangulations

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

def f1(x,y):
  return x**2+y**2

x=np.random.uniform(low=-3,high=3,size=(100))
y=np.random.uniform(low=-3,high=3,size=(100))
z=f1(x,y)
ax=plt.axes(projection='3d')
ax.plot_trisurf(x,y,z,cmap='viridis')
```
